```python
import random
import pandas as pd

# Generate sample data
data = {
    "College": ["Sri sivasubramaniya nadar college", "Panimalar", "Sairam college", "velammal", "SRM"],
    "Science Courses": [14,5,13,10,11],
    "Commerce Courses": [0,11,12,6,3],
    "Humanities Courses": [5,2,4,21,1]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Save the data to a CSV file
df.to_csv("university_courses.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from the CSV file
df = pd.read_csv("university_courses.csv")

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(df["College"], df["Science Courses"], label="Science")
plt.bar(df["College"], df["Commerce Courses"], label="Commerce", bottom=df["Science Courses"])
plt.bar(df["College"], df["Humanities Courses"], label="Humanities", bottom=df["Science Courses"] + df["Commerce Courses"])

# Customize the plot
plt.xlabel("Colleges")
plt.ylabel("Number of Courses")
plt.title("Number of Courses by College and Discipline")
plt.legend()



plt.show()
```
